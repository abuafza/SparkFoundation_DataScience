```
plt.figure(figsize=(15,15))
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```
import seaborn as sns
# finding the pairwise correlations between the columns and visualising using heatmaps
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

```
plt.rcParams['figure.figsize']=15,11
sns.countplot(x=dataset['iyear'])
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Year')
plt.title('Increase in Terror',size=25, fontweight="bold")
```

```
plt.rcParams['figure.figsize']=15,11
sns.countplot(x=dataset['region_txt'], hue='success', data=dataset)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
```

```
plt.rcParams['figure.figsize']=15,11
sns.barplot(x=dataset['country_txt'].value_counts()[:20].index,y=dataset['country_txt'].value_counts()[:20].values)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
plt.title('Country wise Attack',size=25, fontweight="bold")
```
